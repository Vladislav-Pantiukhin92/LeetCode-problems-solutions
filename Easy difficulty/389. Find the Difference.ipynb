### 389. Find the Difference

```python
"""
You are given two strings s and t.
String t is generated by random shuffling string s and then add one more letter at a random position.
Return the letter that was added to t.
Example 1:

Input: s = "abcd", t = "abcde"
Output: "e"
Explanation: 'e' is the letter that was added.
Example 2:

Input: s = "", t = "y"
Output: "y"
"""

class Solution:
    def findTheDifference(self, s: str, t: str) -> str:
        if not s and not t: # If 's' is an empty string and 't' is an empty string, "" is logically returned
            return ""
        elif not s: # If 's' is an empty string, we return "t"
            return t
        elif not t: # If 't' is an empty string, we return 's'
            return s
        else:
            # It is necessary to sort both the first and the second strings
            # As a result of that, the combination of characters should be the same, except the only letter,
            # which will be eventually returned as the final output
            i = 0 
            j = 0 
            sort_a = sorted(s)
            sort_b = sorted(t)
            while i < len(sort_a) and j < len(sort_b):
                if sort_a[i] == sort_b[j]:
                    i += 1
                    j += 1
                else:
                    return sort_b[j]
            return sort_b[j]
        
a = Solution()
a.findTheDifference("abcd", "abcde")
```
