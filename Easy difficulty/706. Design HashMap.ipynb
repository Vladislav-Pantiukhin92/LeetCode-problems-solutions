### 706. Design HashMap

```python
"""
Design a HashMap without using any built-in hash table libraries.
Implement the MyHashMap class:

MyHashMap() initializes the object with an empty map.
void put(int key, int value) inserts a (key, value) pair into the HashMap. If the key already exists in the map, update the corresponding value.
int get(int key) returns the value to which the specified key is mapped, or -1 if this map contains no mapping for the key.
void remove(key) removes the key and its corresponding value if the map contains the mapping for the key.
Example 1:

Input
["MyHashMap", "put", "put", "get", "get", "put", "get", "remove", "get"]
[[], [1, 1], [2, 2], [1], [3], [2, 1], [2], [2], [2]]
Output
[null, null, null, 1, -1, null, 1, null, -1]

Explanation
MyHashMap myHashMap = new MyHashMap();
myHashMap.put(1, 1); // The map is now [[1,1]]
myHashMap.put(2, 2); // The map is now [[1,1], [2,2]]
myHashMap.get(1);    // return 1, The map is now [[1,1], [2,2]]
myHashMap.get(3);    // return -1 (i.e., not found), The map is now [[1,1], [2,2]]
myHashMap.put(2, 1); // The map is now [[1,1], [2,1]] (i.e., update the existing value)
myHashMap.get(2);    // return 1, The map is now [[1,1], [2,1]]
myHashMap.remove(2); // remove the mapping for 2, The map is now [[1,1]]
myHashMap.get(2);    // return -1 (i.e., not found), The map is now [[1,1]]
"""
# Implementation of the HashMap data structure:
class MyHashMap:

    def __init__(self):
        self.size = 20 # The size is 20, so the array will initially consist of 20 None elements
        self.array = self.size * [None] # The initial None array

    def put(self, key: int, value: int) -> None: # The method adds elements to the HashMap
        # Let's elaborate the hash_address:
        h = 0
        for i in str(key):
            h += ord(i)
        hash_index = h % self.size
# If the hash_address is vacant, then, we just add the key-value pair to create the nested list        
        if self.array[hash_index] is None: 
            self.array[hash_index] = []
            self.array[hash_index].append([key, value])
        else:
# If the key is already there, we just update the value, if necessary
            for i in self.array[hash_index]:
                if i[0] == key:
                    i[1] = value
                    return 
            self.array[hash_index].append([key, value])
# To get element we elaborate the hash_address again
    def get(self, key: int) -> int:
        h = 0
        for i in str(key):
            h += ord(i)
        hash_index = h % self.size
        if self.array[hash_index] is None: # If None - the key doesn't exist - return -1
            return -1
        else:
            for i in self.array[hash_index]:
                if i[0] == key: # If the key is found - return the key
                    return i[1]
            return -1 # If not None but the key was not found as well - return -1

    def remove(self, key: int) -> None:
        h = 0 # Again the hash_address should be elaborated
        for i in str(key):
            h += ord(i)
        hash_index = h % self.size
        if self.array[hash_index] is None: # If None, it means that nothing can be removed
            return
        else:
            for i in range(len(self.array[hash_index])):
                if self.array[hash_index]:
                    try: 
                        if self.array[hash_index][i]: # If the key is found - we remove the entire key-value pair
                            if self.array[hash_index][i][0] == key:
                                self.array[hash_index].pop(i)
                    except IndexError: # If the error is raised - then nothing can be removed
                        return 
```
